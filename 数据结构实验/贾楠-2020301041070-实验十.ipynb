```python
class ListItem(object):
    def __init__(self, key, value) -> None:
        self.key = key
        self.value = value


class SScquenceList(object):
    # 初始化顺序表函数
    def __init__(self) -> None:
        self.SeqList = []
    # 创建顺序表函数

    def CreateSequenceListByInput(self):
        print("----------------------")
        print("请输入数据后按回车键确认，若想结束输入请按“#”。")
        print('----------------------')
        self.SeqList.append(ListItem(int(0), 0))
        i = 1
        a = input('请输入元素：')
        while a != '#':
            self.SeqList.append(ListItem(int(a), i))
            i += 1
            a = input('请输入元素：')
        print('顺序表SL创建完成')
    # 遍历顺序表函数

    def TraverseElementSet(self):
        for i in range(1, len(self.SeqList)):
            print(self.SeqList[i], end='')

    # 直接插入排序
    def InsertSort(self):
        SeqListLen = len(self.SeqList)
        for i in range(2, SeqListLen):
            self.SeqList[0].key = self.SeqList[i].key
            index = i
            while self.SeqList[index-1].key > self.SeqList[0].key:
                self.SeqList[index].key = self.SeqList[index-1].key
                index -= 1
            self.SeqList[index].key = self.SeqList[0].key

    # 冒泡排序
    def BubbleSort(self):
        SeqListLen = len(self.SeqList)
        for i in range(1, SeqListLen-1):
            for j in range(1, SeqListLen-i):
                if self.SeqList[j+1].key < self.SeqList[j].key:
                    self.SeqList[0].key = self.SeqList[j].key
                    self.SeqList[j].key = self.SeqList[j+1].key
                    self.SeqList[j+1].key = self.SeqList[0].key
    # 输出函数
    def printout(self):
        print('**********************')
        print('\n(1)初始化顺序表SL：',end='')
        try:
            self.__init__()
            print("顺序表SL初始化成功")
        except:
            print("顺序表SL初始化失败")
        print('\n(2)创建顺序表SL')
        try:
            self.CreateSequenceListByInput()
        except ValueError:
            print('输入有误，请重新输入：')
            self.CreateSequenceListByInput()
        print('\n(3)执行直接插入排序算法')
        self.InsertSort()
        print('排序结果为：')
        for item in self.SeqList[1:]:
            print(item.key,end=' ')
        print('\n(4)执行冒泡排序算法')
        self.BubbleSort()
        print('排序结果为：')
        for item in self.SeqList[1:]:
            print(item.key,end=' ')
if __name__ =='__main__':
    SL=SScquenceList()
    SL.printout()
```
