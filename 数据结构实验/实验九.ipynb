```python
class StaticTableElement(object):
    def __init__(self,key) -> None:
        self.key=key

#定义一个静态查找表
class StaticTable(object):
    def __init__(self) -> None:
        self.data=[]
        self.length=0
    #判断静态查找表为空的函数
    def InEmpty(self):
        if self.length==0:
            print("静态查找表SST为空！")
            return 1
        else:
            print("静态查找表SST不为空！",self.data)
    #创建静态查找表函数
    def CreateStaticTable(self):
        print("--------------------")
        print("请输入数据后按回车键确认，若想结束输入请按#")
        Element=input("请输入元素：")
        while Element!='#':
            self.data.append(StaticTableElement(int(Element)))
            Element=input("请输入元素：")
        self.length=len(self.data)
        print("静态查找表SST创建完成")
    #获取静态查找表长度的函数
    def GetLength(self):
        print("静态查找表SST中元素的个数为",self.length,"个")

    
        
    
    #顺序查找函数
    def SequenceSearch(self):
        key=int(input("请输入想要查找的关键字："))
        ipos=-1
        for i in range(self.length):
            if self.data[i].key==key:
                ipos=i
                break
        if ipos is -1:
            print("查找失败！当前静态查找表中不存在关键字为",key,"的元素")
        else:
            print("查找成功！关键字为",self.data[ipos].key,"的元素，位于当前静态查找表的第",ipos+1,"个位置")
            
    #折半查找函数
    def BinarySearch(self):
        key=int(input("请输入想要查找的关键字："))
        low=0
        high=self.length-1
        while low<=high:
            mid=int((low+high)/2)
            if key is self.data[mid].key:
                return mid and print("查找成功！关键字为",self.data[mid].key,"的元素，位于当前静态查找表的第",mid+1,"个位置")
            elif key<self.data[mid].key:
                high=mid-1
            elif key>self.data[mid].key:
                low=mid+1
        return False and print("查找失败！当前静态查找表中不存在关键字为",key,"的元素")
    #遍历静态查找表函数
    def TraverseElement(self):
        print("----------遍历静态查找表中元素----------")
        for i in range(self.length):
            print("第",i+1,"个元素的值为",self.data[i].key)
        print()
    #输出函数
    def PrintOut(self):
        print("**************实现顺序查找的算法****************")
        print("\n(1)初始化静态查找表SST:",end="")
        try:
            self.__init__()
            print("静态查找表SST初始化成功")
        except:
            print("静态查找表SST初始化失败")
        print("\n(2)判断静态查找表SST是否为空：",end="")
        self.InEmpty()

        print("\n(3)创建静态查找表SST")
        try:
            self.CreateStaticTable()
        except ValueError:
            print("输入有误，请重新输入！")
            self.CreateStaticTable()    

        print("\n(4)输出静态查找表SST")
        try:
            self.TraverseElement()
        except:
            print("输出静态查找表SST出错！")

        print("\n(5)采用顺序查找算法查找关键字为9的数据元素")
        try:
            self.SequenceSearch()
        except:
            print("查找关键字为9的数据元素出错！")

        print("****************实现折半查找的算法****************")
        print("\n(6)采用折半查找算法查找关键字为17的数据元素")
        try:
            self.BinarySearch()
        except:
            print("查找关键字为17的数据元素出错！")
        print("***********************************************")
if __name__=='__main__':
    SST=StaticTable()
    SST.PrintOut()

        
```
